```python
## loading the datasets
covid_2020 = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties-2020.csv')
```

```python
## drop unrevelant columns
covid_cases = covid_2020[['date', 'county', 'state', 'cases']]

## remove unnecessary states, Missouri is included because it has missing value in population dataset
print(covid_cases.state.unique())
states_to_drop = ['Missouri', 'Northern Mariana Islands', 'Puerto Rico', 'Virgin Islands', 'Guam', 'District of Columbia']
covid_cases = covid_cases[~covid_cases.state.isin(states_to_drop)]
print(covid_cases.state.nunique())
```

```python
## filter dates to correspond with population dataframe
covid = covid_cases[covid_cases['date'] <= '2020-12-04']
```

```python
# population = population[['total_population']]
population = general[['date', 'state', 'county', 'total_population']]
```

```python
## check if there is any null value
print(population.isna().any())
nan_rows = population[population['total_population'].isnull()]
print(nan_rows.state.unique())
print(population.state.nunique())
## it turns out that all missing values are from ['Missouri' 'Northern Mariana Islands' 'Puerto Rico' 'Virgin Islands'], we would simply drop them along with one other sovereign state Guam
states_to_drop = ['Missouri', 'Northern Mariana Islands', 'Puerto Rico', 'Virgin Islands', 'Guam', 'District of Columbia']
population = population[~population.state.isin(states_to_drop)]
population = population.rename(columns={'total_population': 'population'})
print(population.state.nunique())
```

```python
education = pd.read_csv('/content/drive/MyDrive/108/education.txt', sep='	')
education.head()
```

```python
education = education.rename(columns={'Rank': 'Edu_Index', 'Average Education Index County / Population': 'County/Population'})

## drop irrelevant information (population and state) to match with dataframes above
education['state'] = education['County/Population'].str.split(",").str[1].str.split("/").str[0]
education['county'] = education['County/Population'].str.split(",").str[0]
education = education.drop('County/Population', axis=1)

## drop states corresponding with other datasets
states_to_drop = [' MO ', ' MP ', ' PR ', ' VI ', ' GU ', ' DC ']
education = education[~education.state.isin(states_to_drop)]
print(education.state.nunique())
```

```python
states_abbrev = {
    'AK': 'Alaska',
    'AL': 'Alabama',
    'AR': 'Arkansas',
    'AZ': 'Arizona',
    'CA': 'California',
    'CO': 'Colorado',
    'CT': 'Connecticut',
    'DC': 'District of Columbia',
    'DE': 'Delaware',
    'FL': 'Florida',
    'GA': 'Georgia',
    'HI': 'Hawaii',
    'IA': 'Iowa',
    'ID': 'Idaho',
    'IL': 'Illinois',
    'IN': 'Indiana',
    'KS': 'Kansas',
    'KY': 'Kentucky',
    'LA': 'Louisiana',
    'MA': 'Massachusetts',
    'MD': 'Maryland',
    'ME': 'Maine',
    'MI': 'Michigan',
    'MN': 'Minnesota',
    'MO': 'Missouri',
    'MS': 'Mississippi',
    'MT': 'Montana',
    'NC': 'North Carolina',
    'ND': 'North Dakota',
    'NE': 'Nebraska',
    'NH': 'New Hampshire',
    'NJ': 'New Jersey',
    'NM': 'New Mexico',
    'NV': 'Nevada',
    'NY': 'New York',
    'OH': 'Ohio',
    'OK': 'Oklahoma',
    'OR': 'Oregon',
    'PA': 'Pennsylvania',
    'RI': 'Rhode Island',
    'SC': 'South Carolina',
    'SD': 'South Dakota',
    'TN': 'Tennessee',
    'TX': 'Texas',
    'UT': 'Utah',
    'VA': 'Virginia',
    'VT': 'Vermont',
    'WA': 'Washington',
    'WI': 'Wisconsin',
    'WV': 'West Virginia',
    'WY': 'Wyoming'
}
def standardized_state(str):
    str = str.strip()
    return states_abbrev[str]

def standardized_county(str):
    str = str.replace("County", "")
    str = str.replace("City", "")
    str = str.replace("Parish", "")
    return str
```

```python
education['state'] = education['state'].apply(standardized_state)
education['county'] = education['county'].apply(standardized_county)
education.head()
```

```python
covid_cases['county'] = covid_cases['county'].apply(standardized_county)
covid_cases.head()
```

```python
population['county'] = population['county'].apply(standardized_county)
population.head()
```

```python
## we keep the most recent date in both dataset
print(population.date.max())
print(covid.date.max())
population_recent = population[population['date'] == "2020-12-04"].reset_index().drop(columns=['date'])
covid_recent = covid[covid['date'] == "2020-12-04"].reset_index().drop(columns=['date'])

## get rid of missing county
# common_county = set(covid_recent['county'].unique().tolist()).intersection(set(population_recent['county'].unique().tolist()))
# population_common = population_recent[population_recent['county'].isin(common_county)].drop(columns=['state','date', 'index']).reset_index()
# covid_common = covid_recent[covid_recent['county'].isin(common_county)].drop(columns=['state','date', 'index']).reset_index()

## remove duplicate county
# population_common = population_common.drop_duplicates(subset='county', keep='first')
# covid_common = covid_common.drop_duplicates(subset='county', keep='first')

## merge population and covid cases based on county
joint_pop_covid = pd.merge(population_recent, covid_recent, on=['state','county']).drop(columns=['index_x', 'index_y'])
# joint.isna().sum()
joint_pop_covid

```

```python
# education_common = education[education['county'].isin(common_county)].reset_index().drop_duplicates(subset='county')

joint = pd.merge(joint_pop_covid, education, on=['state','county'])
joint
```

```python
import patsy
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import ttest_ind, chisquare, normaltest
```

```python
## attaching the rate of covid positive cases within each county with respect to each county's population
joint["covid_positive_rate"] = joint_pop_covid.cases / joint_pop_covid.population
joint
```

```python
joint[joint['state']=='Alabama']['cases'].sum()
```

```python
test = joint.groupby('state')['cases','population'].sum()
test['covid_rate'] = test.cases / test.population
test2 = joint.groupby('state')['Edu_Index'].mean()
mer = pd.merge(test,test2, on='state').reset_index()
mer
```

```python
plt.figure(figsize=(12,8))
plt.xlabel("Edu_Index", fontsize=18)
plt.ylabel("covid_positive_rate", fontsize=18)
plt.suptitle("Joint distribution of Edu_Index vs covid_positive_rate", fontsize= 20)
x = mer.Edu_Index
y = mer.covid_rate
plt.plot(x, y, 'bo', alpha=0.2)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = 'red')
plt.show()
```

```python
outcome, predictors = patsy.dmatrices('covid_rate ~ Edu_Index', mer)
model = sm.OLS(outcome,predictors)
res_1 = model.fit()
print(res_1.summary())
```

```python
## dealing with outliers, remove when covid_rate > 0.08 
mer2 = mer[mer['covid_rate']<0.08]
plt.figure(figsize=(12,8))
plt.xlabel("Edu_Index", fontsize=18)
plt.ylabel("covid_positive_rate", fontsize=18)
plt.suptitle("Joint distribution of Edu_Index vs covid_positive_rate", fontsize= 20)
x = mer2.Edu_Index
y = mer2.covid_rate
plt.plot(x, y, 'bo', alpha=0.2)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = 'red')
plt.show()
```

```python
outcome, predictors = patsy.dmatrices('covid_rate ~ Edu_Index', mer2)
model = sm.OLS(outcome,predictors)
res_2 = model.fit()
print(res_2.summary())
```
